```python
class SGD:
    
    def __init__(self,K=10,lr=0.002,lamda=0.02, steps=10, error_func=None):
        self.K = K
        self.lr = lr
        self.lamda = lamda 
        self.steps = steps
        self.error = SGD.error if not error_func else error_func
        
    def __initPQ__(self, shape):
        M,N = shape
        P = np.random.normal(scale= 1./self.K,size = (M,self.K))
        Q = np.random.normal(scale= 1./self.K,size = (self.K,N))
        return P,Q
    
    @staticmethod
    def error(R,P,Q,lamda = 0.02):
        data = R.data
        rows = R.row
        cols = R.col
        e = 0
        errors=[]
        for z in range(len(data)):

            rxi = data[z]
            j = rows[z]
            k = cols[z]
            Aij = np.dot(P[j,:],Q[:,k])
            length_term = [norm(P[j,:]),norm(Q[:,k])]
            if rxi>0:
                e = e+(rxi-Aij)**2 +lamda*(sum([i**2 for i in length_term])) 
                errors.append(e)
        return e

    def train(self,R,lr=0.002,lamda=0.02, steps=10):
        P,Q = self.__initPQ__(R.shape)

        params = [R,P,Q,self.lamda]
        rmse = np.sqrt(self.error(*params)/len(R.data))
        print(f"Initial RMSE {'{:.4f}'.format(rmse)}")

        for step in range(self.steps):
            for z in range(len(R.data)):
                rxi = R.data[z]
                j = R.row[z]
                k = R.col[z]
                if rxi>0:
                    eui = rxi-np.dot(P[j,:],Q[:,k])
                    #update 
                    P[j,:]=P[j,:]+2*lr*(eui*Q[:,k]-lamda*P[j,:])  
                    Q[:,k]=Q[:,k]+2*lr*(eui*P[j,:]-lamda*Q[:,k])

            rmse = np.sqrt(self.error(*params)/len(R.data))
            print(f"RMSE {'{:.4f}'.format(rmse)}")
        print(f"Final RMSE {'{:.4f}'.format(rmse)}")
        return P,Q
        
```

```python
class LFM:
    
    def __init__(self, rating_matrix, optimizer):
        self.R = rating_matrix
        self.optimizer = optimizer

    def train(self):
        
        P,Q = self.optimizer.train(self.R)
        R_hat = np.dot(P,Q)
        return R_hat, P, Q
    
    def GetUserWatchedContent( user_id, return_rating = False):
        userdata = rating[rating.user_id == user_id]
        if return_rating:
            return {userdata.iloc[i].movie : userdata.iloc[i].movie for i in range(len(userdata))}
        else:
            return list(userdata['movie'])
    
    @staticmethod
    def GetPredictions(df_R_hat, top_n, verbose = False):

        print("Enter the User Id")
        userid = int(input())
        if not df_R_hat.index.isin([userid]).any():
            print(f'User {userid} does exist in  matrix')
            return None

        pred_df = df_R_hat[ df_R_hat.index == userid].T.sort_values(by = userid, ascending = False)
        old_content = LFM.GetUserWatchedContent(userid, return_rating = True)
        top_content = pred_df[ ~pred_df.index.isin(old_content)][:top_n].index

        if verbose:
            print(f'User {userid} already watched:\n ###---------------------####')
            for content in old_content:


                print(f'{str(old_content[content])}')
            print(f'\nWe recommend:\n ###---------------------###')
            for content in top_content:
                print(content)

```

```python
from scipy.sparse import coo_matrix


R = coo_matrix(r_matrix.values)
optimizer = SGD()
lfm = LFM(rating_matrix=R, optimizer=optimizer)
R_hat,P_hat,Q_hat = lfm.train()
df_R_hat = pd.DataFrame(data = R_hat, index = r_matrix.index, columns= r_matrix.columns)
```
