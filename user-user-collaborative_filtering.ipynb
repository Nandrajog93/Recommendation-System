#### ` User-Based Collaborative Filtering `

`“User-based collaborative filters do`: They find users similar to a particular user and then recommend products that those users have liked to the first User.”



`Build a matrix using pivot tabel`

```python
r_matrix = X_train.pivot_table(values='rating',index = 'user_id',columns = 'movie')
```

```python
r_matrix.T
```

`If we see the above cell we can observed clearly that each  row contains user and every column represent a particular movie rating.
& we can also observe that matrix is soo sparse`

##### `First approach`

` “Let's first build one of the simplest collaborative filters possible. This simply takes in user_id and movie_id and outputs the mean rating for the movie by all the users who have rated it. No distinction is made between the users. In other words, the rating of each user is assigned equal weight.`

`It is possible that some movies are available only in the test set and not the training set (and consequentially, not in our ratings matrix). In such cases, we will just default to a rating of 3.0, like the baseline model:” `



` Mean-Rating collaborative filltering `

```python
def cf_user_mean(user_id,movie_id):
    
    #Check if movie exist in the r_matix
    
    if movie_id in r_matrix:
        #If movie present in r_matrix compute the mean of all the raiting givien to the movie
        
        mean_rating = r_matrix[movie_id].mean()
    else:
        mean_rating=3.0
        
    return mean_rating
```

```python
score(cf_user_mean)
```

### `Weighted Mean`

` Here we are using `

` “coefficient will be one of the similarity metrics that we computed ”`



##### Fill NA ==0 
##### cosine similarities does not work with NAN 

```python
r_matrix_dummy = r_matrix.fillna(0)
```

```python
r_matrix_dummy.head()
```

```python
r_matrix.head()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
cosine_simi = cosine_similarity(r_matrix_dummy,r_matrix_dummy)
```

```python
cosine_similarity_df = pd.DataFrame(cosine_simi,index=r_matrix_dummy.index,
                                   columns=r_matrix_dummy.index)
```

```python
cosine_similarity_df
```

```python
def cf_users_wmean(user_id, movie_id):
# for movie_id in r_matrix: 
#     print(movie_id)
    if movie_id in r_matrix:
        # Get the similarity score
        sim_scores = cosine_similarity_df[user_id]
    #         print(sim_score)

        #Get the user ratings for the movie

        m_rating = r_matrix[movie_id]
#         print(m_rating)
        
        #Extract the indices containing NaN in the rating series
        
        idx = m_rating[m_rating.isnull()].index
#         print(idx)
        #Drop NaN 
        
        m_rating = m_rating.dropna()
        #Drop cosine from corresponding
        sim_score = sim_scores.drop(idx)
#         print(sim_score)
        
        #Compute the final weighted mean
        
        wmean_rating = np.dot(sim_score,m_rating)/sim_score.sum()
#         print(wmean_rating)
    
    else:
        wmean_rating = 3.0
        
    return wmean_rating
        
```

```python
score(cf_users_wmean)
```

#### `Gender Based Collaborative filttering`

```python
def cf_gender(user_id,movie_id):
    
    #Check if movie is exist in the matrix
    if movie_id in r_matrix:
        #Identify the gender of the user
        gender = user.loc[user_id]['sex']
#         print(gender)
        #compute the mean rating given by that gender of the movie
        if gender in gender_mean[movie_id]:
            gender_rating = gender_mean[movie_id][gender]
        else:
            gender_rating = 3.0
    else:
        gender_rating = 3.0

    return gender_rating
        
        
```

```python
score(cf_gender)
```
