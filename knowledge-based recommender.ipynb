The metric 

V = No. of votes generated by the movies 

M = No. of minimum votes required for a movie to qualify 

R = The mean rating of the movie 

C = The mean rating of the whole dataset

The prerequisties

let's calculate the value of m 

```python
m = read['vote_count'].quantile(0.80)
```

it turn out the value of m is 50.0

We can see the only 20% of the movies have gained more then 50 votes

Another prerequistie that we want movies with above 45 min runtime and shorter then 300 

```python
q_movies = read[(read['runtime'] >=45) & ( read['runtime']<= 300)]
```

```python
# q_movies = df[(df['runtime'] >= 45) & (df['runtime'] <= 300)]
```

```python
movies = q_movies[q_movies['vote_count']>=m]
```

```python
movies.shape
```

```python
type(movies['vote_count'])
```

Calculating the Score

Calculate the C 

```python
C = movies['vote_average'].mean()
```

```python
C
```

```python
# Function to compute the IMDB weighted rating for each movie
def weighted_rating(x, m=m, C=C):
    v = x['vote_count']
    R = x['vote_average']
    # Compute the weighted score
    return (v/(v+m) * R) + (m/(m+v) * C)
```

Now we see the average rating of the movie is 6.35 

```python
movies['vote_average']
```

```python
movies['score'] = movies.apply(weighted_rating, axis=1)
```

```python
movies
```

```python
#Sort the movies 
```

```python
movies = movies.sort_values('score', ascending=False)
```

```python
def build_chart(df,percetile):
    
     #Ask for preferred genres”
    print("Input a preferred genres") 
    genres = input()

    # Ask for lower limit of duration
    
    print("Input the lower duration")
    low_time = int(input())
    
    # Ask for upper limit of duration
    
    print("Input the Upper duration ")
    upper_limit = int(input())
    
    
    # Ask for the earliest year

    print("Input the earliest year")
    earliest_year = int(input())
    
    # Ask for the latest year
    print("Input the latest year ")
    latest_year = int(input())
    
    #Define a new movies variable to store the preferred movies. Copy the contents of gen_df to movies”
    
    movies = gen_df.copy()
    
    #filter movies on the basis of the above conditions
    
    movies = movies[(movies['genre'] == genres)& 
                    (movies['runtime'] >=low_time) & 
                    (movies['runtime']<= upper_limit) & 
                    (movies['year'] >= earliest_year) &
                    (movies['year']<= latest_year)]
    
    # Compute the value of C & M
    
    C = movies['vote_average'].mean()
    m = movies['vote_count'].quantile(percetile)
    
    #Only consider movies that have higher than m votes. Save this in a new dataframe q_movies”
    
    q_movies = movies.copy().loc[movies['vote_count'] >=m]
    
    q_movies['score'] = q_movies.apply(lambda x: (x['vote_count']/(x['vote_count']+m) * x['vote_average'])  + (m/(m+x['vote_count']) * C),axis=1)
#     print(q_movies)
    
    q_movies = q_movies.sort_values('score',ascending = False)
    
    return q_movies

#     return q_movies

    
    
```

```python
build_chart(gen_df,percetile=0.9)
```
