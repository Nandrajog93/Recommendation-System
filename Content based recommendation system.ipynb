```python
# def converstion(x):
    
#     lower_ =[]
#     if isinstance(x,list):
#         for i in x:
#             lower_.append(str.lower(i.replace(" ", " ")))
#         return lower_
            
#     if isinstance(x,str):
#         lower_.append(str.lower(x.replace(" ", " ")))
#         return lower_
#     else: 
#         return ''

            
            
            
# Function to sanitize data to prevent ambiguity. It removes spaces and converts to lowercase
def converstion(x):
    if isinstance(x, list):
        #Strip spaces and convert to lowercase
        return [str.lower(i.replace(" ", "")) for i in x]
    else:
        #Check if director exists. If not, return empty string
        if isinstance(x, str):
            return str.lower(x.replace(" ", ""))
        else:
            return ''
```

```python
columns_ = ['cast', 'Director','keywords', 'genres']
```

```python
for features in clean_data[columns_]:
    clean_data[features]  = clean_data[features].apply(converstion)
```

```python
def create_soup(x):
    return ' '.join(x['keywords']) + ' ' + ' '.join(x['cast']) + ' ' + x['Director'] + ' ' + ' '.join(x['genres'])
```

```python
clean_data['soup'] = clean_data.apply(create_soup,axis=1)
```

```python
from sklearn.feature_extraction.text import CountVectorizer


count_ = CountVectorizer(encoding='english')
count_matrix = count_.fit_transform(clean_data["soup"])
```

```python
from sklearn.metrics.pairwise import cosine_similarity

cosine_sim = cosine_similarity(count_matrix,count_matrix)
```

```python
cosine_sim.shape
```

```python
indices = pd.Series(clean_data.index,index=clean_data['title']).drop_duplicates()
```

```python
def meta_recommendation(title,data,cosine_sim,indicies):
    
    idx = indices[title]
    
    score = list(enumerate(cosine_sim[indices[idx]]))
    
    score = sorted(score,key=lambda x:x[1], reverse=True )
    
    score = score[1:11]
    
    movies_recommendation =[]
    for i in score:
        movies_recommendation.append(i[0])
    
    return data.iloc[movies_recommendation]['title']
        
    
    
```

```python
meta_recommendation(title='Jumanji',data=clean_data,cosine_sim=cosine_sim,indicies=indices)
```
